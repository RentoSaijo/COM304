<h3 style="display:inline">(1)</h3> Use induction to show that: $\displaystyle\sum_{k=1}^{n+a}k=\frac{n(n+1)}{2}+an+\frac{a(a+1)}{2}.$

We prove by induction on $n$ (with $a$ as some constant) that
$$
\sum_{k=1}^{n+a}k=\frac{n(n+1)}{2}+an+\frac{a(a+1)}{2}.
$$

Base case ($n=1$): We compute both sides and show they are equal.

Left-hand side:
\begin{align*}
\sum_{k=1}^{1+a}k
&=\frac{(a+1)(a+2)}{2}.
\end{align*}

Right-hand side:
\begin{align*}
\frac{1(1+1)}{2}+a(1)+\frac{a(a+1)}{2}
&=1+a+\frac{a(a+1)}{2}\\
&=\frac{2+2a+a(a+1)}{2}\\
&=\frac{2+2a+a^2+a}{2}\\
&=\frac{a^2+3a+2}{2}\\
&=\frac{(a+1)(a+2)}{2}.
\end{align*}

Since both sides equal $\frac{(a+1)(a+2)}{2}$, the statement holds for $n=1$.

Inductive step: Assume for some $n\ge 1$ that
$$
\sum_{k=1}^{n+a}k=\frac{n(n+1)}{2}+an+\frac{a(a+1)}{2}.
$$
We show the statement holds for $n+1$ by computing both sides and showing they match.

Left-hand side:
\begin{align*}
\sum_{k=1}^{n+1+a}k
&=\sum_{k=1}^{n+a}k+(n+a+1).
\end{align*}
Using the induction hypothesis,
\begin{align*}
\sum_{k=1}^{n+1+a}k
&=\left(\frac{n(n+1)}{2}+an+\frac{a(a+1)}{2}\right)+(n+a+1)\\
&=\frac{n(n+1)}{2}+an+n+a+1+\frac{a(a+1)}{2}\\
&=\frac{n(n+1)+2n+2}{2}+a(n+1)+\frac{a(a+1)}{2}\\
&=\frac{(n+1)(n+2)}{2}+a(n+1)+\frac{a(a+1)}{2}.
\end{align*}

Right-hand side (with $n$ replaced by $n+1$):
\begin{align*}
\frac{(n+1)\big((n+1)+1\big)}{2}+a(n+1)+\frac{a(a+1)}{2}
&=\frac{(n+1)(n+2)}{2}+a(n+1)+\frac{a(a+1)}{2}.
\end{align*}

The left-hand side and right-hand side are the same expression, so the statement holds for $n+1$.

Therefore, by induction, for all $n\ge 1$,
$$
\sum_{k=1}^{n+a}k=\frac{n(n+1)}{2}+an+\frac{a(a+1)}{2}.
$$

<h3 style="display:inline">(2)</h3> Using the proven equation above, put an upper bound on the sum (could use big-oh): $\displaystyle\sum_{k=1}^{n+a}k.$

Using the identity from (1),
$$
\sum_{k=1}^{n+a}k=\frac{n(n+1)}{2}+an+\frac{a(a+1)}{2}.
$$
Assume $a\ge 0$ is a constant and $n\ge a+1$. Then $n+a\le 2n$, so
\begin{align*}
\sum_{k=1}^{n+a}k
&\le \sum_{k=1}^{2n}k
=\frac{(2n)(2n+1)}{2}
=n(2n+1)
\le n(3n)
=3n^2.
\end{align*}
Therefore,
$$
\sum_{k=1}^{n+a}k=O(n^2).
$$

<h3 style="display:inline">(3)</h3> Using the proven equation above, put a lower bound on the sum (could use big-omega): $\displaystyle\sum_{k=1}^{n+a}k.$

Using the identity from (1),
$$
\sum_{k=1}^{n+a}k=\frac{(n+a)(n+a+1)}{2}.
$$
Assume $a\ge 0$ and $n\ge 1$. Then $n+a\ge n$, so
\begin{align*}
\sum_{k=1}^{n+a}k
&\ge \sum_{k=1}^{n}k
=\frac{n(n+1)}{2}
\ge \frac{n^2}{2}. && \text{(Appendix A.1)} \\
\end{align*}
Therefore,
$$
\sum_{k=1}^{n+a}k=\Omega(n^2).
$$

<h3 style="display:inline">(4)</h3> Split the following sum into two parts and then find a closed form or upper bound: $\displaystyle\sum_{k=2}^{2n}k^2$ 

We split the sum as
\begin{align*}
\sum_{k=2}^{2n}k^2
&=\sum_{k=1}^{2n}k^2-\sum_{k=1}^{1}k^2.
\end{align*}
Using the formula $\sum_{k=1}^{m}k^2=\frac{m(m+1)(2m+1)}{6}$ from Appendix A.4 and explicit substition, we get
\begin{align*}
\sum_{k=2}^{2n}k^2
&=\frac{(2n)(2n+1)\big(2(2n)+1\big)}{6}-1^2\\
&=\frac{(2n)(2n+1)(4n+1)}{6}-1\\
&=\frac{n(2n+1)(4n+1)}{3}-1.
\end{align*}
Therefore,
$$
\sum_{k=2}^{2n}k^2=\frac{n(2n+1)(4n+1)}{3}-1.
$$

<h3 style="display:inline">(5*)</h3> Put the sums back together to simply and find the closed form or bound:<br> $3+\displaystyle\sum_{i=3}^{n/2}i + \displaystyle\sum_{k=1}^{a}(n-a+k) + \displaystyle\sum_{j=n/2+1}^{n-a}j.$

Assume $n$ is even and that $0\le a\le n/2$ (so the index ranges make sense). Rewrite the middle sum by a change of variables: let
$$
m=n-a+k.
$$
When $k=1$, $m=n-a+1$, and when $k=a$, $m=n$. Thus,
$$
\sum_{k=1}^{a}(n-a+k)=\sum_{m=n-a+1}^{n}m.
$$
Thus,
\begin{align*}
3+\sum_{i=3}^{n/2}i+\sum_{k=1}^{a}(n-a+k)+\sum_{j=n/2+1}^{n-a}j
&=3+\sum_{i=3}^{n/2}i+\sum_{j=n/2+1}^{n-a}j+\sum_{m=n-a+1}^{n}m.
\end{align*}
The three sums are consecutive pieces:
$$
\sum_{i=3}^{n/2}i
\quad\text{ends at } \frac{n}{2},\qquad
\sum_{j=n/2+1}^{n-a}j
\quad\text{starts at } \frac{n}{2}+1 \text{ and ends at } n-a,
$$
and
$$
\sum_{m=n-a+1}^{n}m
\quad\text{starts at } n-a+1.
$$
So they combine into one sum:
\begin{align*}
\sum_{i=3}^{n/2}i+\sum_{j=n/2+1}^{n-a}j+\sum_{m=n-a+1}^{n}m
&=\sum_{i=3}^{n}i.
\end{align*}
Thus,
\begin{align*}
3+\sum_{i=3}^{n/2}i+\sum_{k=1}^{a}(n-a+k)+\sum_{j=n/2+1}^{n-a}j
&=3+\sum_{i=3}^{n}i.
\end{align*}
Finally, write $3=1+2$ so it can be absorbed into the sum:
\begin{align*}
3+\sum_{i=3}^{n}i
&=1+2+\sum_{i=3}^{n}i
=\sum_{i=1}^{n}i
=\frac{n(n+1)}{2}. && \text{(Appendix A.1)} \\
\end{align*}
