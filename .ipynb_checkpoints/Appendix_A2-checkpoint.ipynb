## Apendix A.2: Bounding Summations

### <u> Math Induction </u>

Lets proove the series sum from A.1:

$\displaystyle\sum_{k=1}^{n}k=\frac{n(n+1)}{2}$

<u>proof:</u>

for $n=1 : \displaystyle\sum_{k=1}^{1}k=\frac{1(1+1)}{2}=1$



assume it holds for $n$, then for $n+1$

$\displaystyle\sum_{k=1}^{n+1}k=\displaystyle\sum_{k=1}^{n}k+(n+1)\;\;$ (by linearity from A.1)

substituting the original RHS then yeilds:

$\displaystyle\sum_{k=1}^{n}k+(n+1)=\frac{n(n+1)}{2}+(n+1)=\frac{(n+1)(n+2)}{2}\;\;\;\blacksquare$

```python
#instance code----------------------------------------------------------------------------------------------
import numpy as np                                                        #import numpy package
n = int(np.random.sample()*1000)+1                                        #random positive n value 1 to 1000
s0,s1 = (n*(n+1))/2,((n+1)*(n+2))/2                                       #close form 

#summation code---------------------------------------------------------------------------------------------
x = 0                                                                     #result starts at 0
for k in range(1,n+1,1): x += k                                           #accumulate each term
print(s0==x)                                                              #show assumption
print(s1==(x+n+1))                                                        #show n+1 holds too
```

### <u>Bounding Terms</u>

For series sums and products we may be able to bound the terms (using limits if that is easier)

$\displaystyle\sum_{k=1}^{n}k\leq\displaystyle\sum_{k=1}^{n}n\;\;$ (since $k=n \implies n\leq n$ )

When $\frac{a_{k+1}}{a_k} \leq r,\forall k \geq 0,\;0<r<1\;\;$:$\;\;\displaystyle\sum_{k=1}^{n}a_k \leq \displaystyle\sum_{k=0}^{\infty}a_0r^k=a_0 \frac{1}{1-r}$

[EX] 

$\displaystyle\sum_{k=1}^{\infty}\frac{k}{3^k}=\displaystyle\sum_{k=0}^{\infty}\frac{k+1}{3^{k+1}}$ if the first term is $a_0=\frac{1}{3}$ and $r \leq \frac{2}{3}$

$\displaystyle\sum_{k=1}^{\infty}\frac{k}{3^k}=\displaystyle\sum_{k=0}^{\infty}\frac{k+1}{3^{k+1}} \leq \left( \frac{1}{3} \right) \left( \frac{1}{1-\frac{2}{3}} \right)$ (by substitution) $= 1$

```python
import math
x = 0
for k in range(1,10000,1): x += (k/pow(3,k))
print(x<=1)
```

### <u>Splitting Summations</u>

Take a series summation or product and break apart the range of the *from* and *to* arguments:


$\displaystyle\sum_{k=1}^{n}k=\displaystyle\sum_{k=1}^{n/2}k+\displaystyle\sum_{k=n/2+1}^{n}k \geq \displaystyle\sum_{k=1}^{n/2}0 + \displaystyle\sum_{k=n/2+1}^{n}\frac{n}{2} = \left( \frac{n}{2} \right)^2 = \Omega(n^2)$ 

and for a constant $k_0>0$:

$\displaystyle\sum_{k=0}^{n}a_k = \displaystyle\sum_{k=0}^{k_0-1}a_k + \displaystyle\sum_{k=0}^{n}a_k = \Theta(1) + \displaystyle\sum_{k=k_0}^{n}a_k$

```python
#instance code----------------------------------------------------------------------------------------------
a = [12,43,12,43,27,19,10,271,281,21,21,43]     #we will use an even sized array
n = len(a)                                      #call its length n

#summation code---------------------------------------------------------------------------------------------
s,s1,s2 = 0,0,0                                 #use full sum s, partial sums s1 and s2
for k in range(1,n):   s  += a[k]               #sum all the elements of a
for k in range(1,n//2): s1 += a[k]              #sum the first half
for k in range(n//2,n): s2 += a[k]              #sum the second half
print(s==s1+s2)                                 #two partial sums do equal a full sum
```

### <u>Approximation by Integrals</u> (advanced)

When we have a monotonically increasing function: $\displaystyle\sum_{k=1}^{n}f(k)$ we can use approximate integration (will be either an over or under estimate of the discrete values):

$\displaystyle\int\limits_{m-1}^{n}f(x)dx \leq \sum_{k=m}^{n}f(k) \leq \int\limits_{m}^{n+1}f(x)dx$

[EX]

$\displaystyle \int\limits_{1}^{n} \frac{dx}{x} = ln(n) \geq \sum_{k=1}^{n} \frac{1}{k} \geq \int\limits_{1}^{n+1} \frac{dx}{x} = ln(n+1) \implies \sum_{k=1}^{n} \frac{1}{k} = \Theta(ln(n))$

```python
#instance code----------------------------------------------------------------------------------------------
import numpy as np
n = 1000                             #use a size of n>1
c = 0.5777155815682065               #set up a constant value

#summation code---------------------------------------------------------------------------------------------
x = 0                                #initialize the summation
for k in range(1,n+1,1): x += (1/k)  #sum 1/k for each k
print(x==(c+np.log(n)))              #check the manual sum verses the closed form bound (add back that constant to get it tight)
```
