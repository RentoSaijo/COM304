## 2.2 Anayzing Algorithms

### preliminary things to consider when conducting a time complexity analysis:

>- assume we have one-core (CPU) and are using the RAM computational model

>- one instruction at a time (no concurrency, or instruction pipelines, etc)

>- each instruction takes one time unit which is what we will count in time complexity analysis

>- data types are integer and float

>- because we are working with inputs of up to size n, we assume each number has clg(n) bits for a constant c>1

### <u>Analysis of Insertion-Sort</u>

We proceed with a time complexity analysis by considering each line of pseudocode as 1-time step (unless we see a function or inner loop)

We are looking to calculate $T(n)$ which will represent the total time complexity of the algorithm. 

We proceed to try to calculate the number of instructions on each line.

Notice that we must execute <b>line 2</b> $n-1$ times for an array of size $n$, but <b>line 5</b> may execute $(n-1)^2$ times

All together, we put all of these estimates from above into a time expression is becomes with the inner loop being denoted by $t$ :


$\displaystyle T(n) = (n-1) + (n-1) + \sum_{j=2}^{n}t_j + \sum_{j=2}^{n}(t_j-1) + \sum_{j=2}^{n}(t_j-1) + (n-1) $

$\displaystyle T(n) = 3(n-1) + \sum_{j=2}^{n}t_j + 2\sum_{j=2}^{n}(t_j-1)$

$\displaystyle T(n) = 3(n-1) + \sum_{j=2}^{n}t_j + 2\sum_{j=2}^{n}t_j + (-2)\sum_{j=2}^{n}1$

$\displaystyle T(n) = 3(n-1) + 3\sum_{j=2}^{n}t_j + (-2)(n-1)$

$\displaystyle T(n) = 3\sum_{j=2}^{n}t_j + 3n - 3 - 2n + 2$

$\displaystyle T(n) = 3\sum_{j=2}^{n}t_j + (n-1)$

Even though the outer loop has the $n-1$ term this may or may not be the leading term depending upon the input instance.

#### <u>Best Case Analysis</u>

The best possible configuration is one that will for the given algorithm cause the fastest run time.

Thinking about how the pseudocode is structured above, we can see the best instance would be when the input is already sorted.

This causes the $t$ term in the sumations to short-circuit early causing 1 time unit:

$\displaystyle T(n) = 3\sum_{j=2}^{n}1 + (n-1)$

$\displaystyle T(n) = 3(n-1) + (n-1)$

$T(n) = 4(n-1) \implies \Omega(n)$

#### <u>Worst Case Analysis</u>

What type of instance would then take the longest?  This one follows from the best case.

We simply put the input instance in the reverse sorted order, this will require the most swaps in the inner loop!

Which in turn cause the $t$ term to eat up to an astonishing $n-1$ time units:

$\displaystyle T(n) = 3\sum_{j=2}^{n}(n-1) + (n-1)$

$\displaystyle T(n) = 3\sum_{j=2}^{n}n + (-3)\sum_{j=2}^{n}1 + (n-1)$

$\displaystyle T(n) = 3\sum_{j=2}^{n}n - 2(n-1)$

$\displaystyle T(n) = 3\sum_{j=1}^{n}n - 3(n-1)$

$\displaystyle T(n) = 3\left( \frac{n(n+1)}{2} \right) + 3(n-1)$

$\displaystyle T(n) = \frac{3}{2}n^2 +\frac{3}{2}n + 3n - 3 \implies O(n^2)$

If we make some updates to our python3 implementation we can check these bounds easily:

```python
#instance code----------------------------------------------------------------------------------------------
import numpy as np                                                        #import numpy package
n = int(np.random.sample()*100)+100                                       #random positive n value 100 to 200
A = np.random.choice([int(np.random.sample()*100)+1 for i in range(n)],n) #now we set each value to 1 to 100

#algorithm code---------------------------------------------------------------------------------------------
def insert_sort(A):
    t_n = 0
    for j in range(1,len(A)):
        k = A[j]
        i = j-1
        t_n += 2
        while i>=0 and A[i]>k:
            A[i+1] = A[i]
            i = i-1
            t_n += 2
        A[i+1] = k
        t_n += 1
    return A,t_n

A_p,t_n = insert_sort(A)
print(n)
print(t_n)
print(n**2)
print(t_n>n and t_n<n**2) #lower bounded by n and upper bounded by n^2
```
