## 3-1

Let
$$
p(n)=\sum_{i=0}^{d}a_i n^i,
$$
where $a_d>0$, and let $k$ be a constant. Use the definitions of asymptotic notation to prove:

a. If $k\ge d$, then $p(n)=O(n^k)$.

b. If $k\le d$, then $p(n)=\Omega(n^k)$.

c. If $k=d$, then $p(n)=\Theta(n^k)$.

### 3-1 (a)

#### My proof:
Assume $k\ge d$. For sufficiently large $n$, $p(n)$ is nonnegative (because the leading term $a_d n^d$ has $a_d>0$ and dominates lower-degree terms). So choose $n_0$ large enough so that for all $n\ge n_0$, we have $0\le p(n)$ and also $n\ge 1$.

Now for $n\ge n_0$ and each $0\le i\le d\le k$, we have $n^i\le n^k$. Therefore,
$$
p(n)\le \sum_{i=0}^{d}|a_i|n^i\le \left(\sum_{i=0}^{d}|a_i|\right)n^k.
$$
Let
$$
c=\sum_{i=0}^{d}|a_i|>0.
$$
Then for all $n\ge n_0$,
$$
0\le p(n)\le c n^k.
$$
Therefore, by the definition of $O(\cdot)$, $p(n)=O(n^k)$.

### 3-1 (b)

#### My proof:
Assuming $k\le d$,
$$
p(n)=a_d n^d+\sum_{i=0}^{d-1}a_i n^i.
$$
Since the second sum has degree at most $d-1$, there exists $n_0$ such that for all $n\ge n_0$,
$$
\sum_{i=0}^{d-1}|a_i|n^i\le \frac{a_d}{2}n^d.
$$
Hence for all $n\ge n_0$,
$$
p(n)
\ge a_d n^d-\sum_{i=0}^{d-1}|a_i|n^i
\ge \frac{a_d}{2}n^d.
$$
Because $k\le d$ and $n\ge 1$, we also have $n^d\ge n^k$, so
$$
p(n)\ge \frac{a_d}{2}n^k.
$$
Let $c=\frac{a_d}{2}>0$. Then for all sufficiently large $n$,
$$
0\le c n^k\le p(n).
$$
Therefore, by the definition of $\Omega(\cdot)$, $p(n)=\Omega(n^k)$.

### 3-1 (c)

#### My proof:
Set $k=d$.

From part (a), we have $p(n)=O(n^d)$. So there exist constants $c_2>0$ and (for sufficiently large $n$)
$$
0\le p(n)\le c_2 n^d.
$$
From part (b), we have $p(n)=\Omega(n^d)$. So there exist constants $c_1>0$ and (for sufficiently large $n$)
$$
0\le c_1 n^d\le p(n).
$$
Choose one $n_0$ large enough so that both inequalities hold for all $n\ge n_0$. Then
$$
0\le c_1 n^d\le p(n)\le c_2 n^d\quad\text{for all }n\ge n_0.
$$

Therefore, by the definition of $\Theta(\cdot)$,
$$
p(n)=\Theta(n^d)=\Theta(n^k).
$$

### 3-2 (e)

$A=n^{\lg c}$, $B=c^{\lg n}$.

#### My answer:
$A=O(B)$: yes

$A=o(B)$: no

$A=\Omega(B)$: yes

$A=\omega(B)$: no

$A=\Theta(B)$: yes

#### My explanation:
First, derive the identity $c^{\lg n}=n^{\lg c}$:
$$
c^{\lg n}
= c^{\frac{\ln n}{\ln 2}}
= e^{\left(\frac{\ln n}{\ln 2}\right)\ln c}
= e^{\left(\frac{\ln c}{\ln 2}\right)\ln n}
= e^{(\lg c)\ln n}
= n^{\lg c}.
$$
So $B=c^{\lg n}=n^{\lg c}=A$ for all $n>0$. Then
$$
\lim_{n\to\infty}\frac{A}{B}=1.
$$
A limit of $1$ means same asymptotic order, so $A=\Theta(B)$, and therefore both $A=O(B)$ and $A=\Omega(B)$.

Because the ratio tends to neither $0$ nor infinity, we have $A\neq o(B)$ and $A\neq\omega(B)$.

Let $f(n)$ and $g(n)$ be asymptotically positive functions. Prove or disprove each conjecture.

### 3-4 (c)

If $f(n)=O(g(n))$, then $\lg f(n)=O(\lg g(n))$, where $\lg g(n)\ge 1$ and $f(n)\ge 1$ for all sufficiently large $n$.

#### My answer:
True.

#### My proof:
Because $f(n)=O(g(n))$, there exist constants $c>0$ and $n_0$ such that
$$
0\le f(n)\le c g(n)\quad\text{for all }n\ge n_0.
$$
For sufficiently large $n$, we are given $f(n)\ge 1$ and $\lg g(n)\ge 1$, so taking logs is valid:
$$
0\le \lg f(n)\le \lg(cg(n))=\lg c+\lg g(n).
$$
Now let
$$
C=|\lg c|+1.
$$
Since $\lg g(n)\ge 1$, we first get:
$$
\lg c+\lg g(n)\le |\lg c|+\lg g(n).
$$
It follows that
$$
|\lg c|+\lg g(n)
=|\lg c|\cdot 1+\lg g(n)
\le |\lg c|\cdot \lg g(n)+\lg g(n)
=(|\lg c|+1)\lg g(n)
=C\lg g(n).
$$
Therefore, for all sufficiently large $n$,
$$
0\le \lg f(n)\le C\lg g(n).
$$
By definition, $\lg f(n)=O(\lg g(n))$.

$f(n)+o(f(n))=\Theta(f(n))$.

#### My answer:
True.

#### My proof:
Take any function $h(n)\in o(f(n))$. By definition of little-$o$, for every constant $c>0$, there exists $n_0$ such that
$$
0\le h(n)<c f(n).
$$
Choose $c=1$. Then, there is some $n_0$ such that for all $n\ge n_0$,
$$
0\le h(n)<f(n).
$$
So for $n\ge n_0$,
$$
f(n)\le f(n)+h(n)<2f(n).
$$
That is,
$$
0\le 1\cdot f(n)\le f(n)+h(n)\le 2\cdot f(n)
$$
for all sufficiently large $n$. By definition of $\Theta$, we get
$$
f(n)+h(n)=\Theta(f(n)).
$$
Since $h(n)$ was an arbitrary element of $o(f(n))$, this proves
$$
f(n)+o(f(n))=\Theta(f(n)).
