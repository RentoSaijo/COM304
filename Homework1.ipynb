<h3 style="display:inline">(1)</h3> Assuming a sequence $a_1, a_2, ... a_n$ and $b_1,b_2,...,b_n$ and $c_1,c_2,...,c_n$ and $d,f,g \in \mathbb{R}$:  $\displaystyle\sum_{i=1}^{n}(da_i^2+fb_i-gc_i)$

Let us use linearity to separate the sum:
$$
\begin{align*}
\sum_{i=1}^{n}(da_i^2+fb_i-gc_i)
&= \sum_{i=1}^{n}da_i^2+\sum_{i=1}^{n}fb_i-\sum_{i=1}^{n}gc_i \\
&= d\sum_{i=1}^{n}a_i^2+f\sum_{i=1}^{n}b_i-g\sum_{i=1}^{n}c_i \\
\end{align*}
$$

Assuming for all $i$ that $0 \leq a_i,b_i,c_i \leq n$ and that $0 \leq d, f, g$, let us bound the sums by selecting the negative and positive sums, respectively (work for if we cannot just assume that $0 \leq d, f, g$ is shown below as well:
$$
\begin{align*}
-g\sum_{i=1}^{n}c_i \leq d\sum_{i=1}^{n}a_i^2+f\sum_{i=1}^{n}b_i-g\sum_{i=1}^{n}c_i \leq d\sum_{i=1}^{n}a_i^2+f\sum_{i=1}^{n}b_i \\
\end{align*}
$$

For each elementary sum, let us find the bounds:
$$
\begin{align*}
\sum_{i=1}^{n}a_i^2
&\leq \sum_{i=1}^{n}n^2 \\
&= n^2(n-1+1) \\
&= n^3 \\
\sum_{i=1}^{n}b_i
&\leq \sum_{i=1}^{n}n \\
&= n(n-1+1) \\
&= n^2 \\
\sum_{i=1}^{n}c_i
&\leq \sum_{i=1}^{n}n \\
&= n(n-1+1) \\
&= n^2 \\
\end{align*}
$$

Let us apply the above bound to the lower bound:
$$
\begin{align*}
-g\sum_{i=1}^{n}c_i &\geq -gn^2 \\
\end{align*}
$$

Let us apply the above bounds to the upper bound:
$$
\begin{align*}
d\sum_{i=1}^{n}a_i^2+f\sum_{i=1}^{n}b_i &\leq dn^3+fn^2 \\
\end{align*}
$$

Putting it altogether, we can bound the original sum:
$$
\begin{align*}
-gn^2 \leq \sum_{i=1}^{n}(da_i^2+fb_i-gc_i) \leq dn^3+fn^2
\end{align*}
$$

### If we cannot assume that $0 \leq d, f, g$:

If $d,f,g$ may be negative, we can no longer keep the sign when bounding the terms. Instead, we take absolute values.

Start from the linearity decomposition:
$$
\begin{align*}
\sum_{i=1}^{n}(da_i^2+fb_i-gc_i) &= d\sum_{i=1}^{n}a_i^2+f\sum_{i=1}^{n}b_i-g\sum_{i=1}^{n}c_i \\
\end{align*}
$$

Take absolute values:
$$
\begin{align*}
\left|\sum_{i=1}^{n}(da_i^2+fb_i-gc_i)\right|
&\leq \left|d\sum_{i=1}^{n}a_i^2\right|+\left|f\sum_{i=1}^{n}b_i\right|+\left|g\sum_{i=1}^{n}c_i\right| \\
&= |d|\sum_{i=1}^{n}a_i^2 + |f|\sum_{i=1}^{n}b_i + |g|\sum_{i=1}^{n}c_i \\
\end{align*}
$$

Using the bounds from above, we get:
$$
\begin{align*}
\left|\sum_{i=1}^{n}(da_i^2+fb_i-gc_i)\right|
&\leq |d|\,n^3 + |f|\,n^2 + |g|\,n^2 \\
&= |d|\,n^3 + (|f|+|g|)n^2 \\
\end{align*}
$$

Therefore,
$$
\sum_{i=1}^{n}(da_i^2+fb_i-gc_i)=O(n^3) \\
$$

<h3 style="display:inline">(9)</h3> give asymptotically tight bounds for $\displaystyle\sum_{k=1}^{n}k^r$

I came up with 2 methods:
### Method 1

Since $k^r$ is increasing for $r\geq 0$, we can bound the sum by bounding each term. For the upper bound, every term satisfies $k^r \leq n^r$, so:
$$
\begin{align*}
\sum_{k=1}^{n}k^r
&\leq \sum_{k=1}^{n}n^r \\
&= n^r(n-1+1) \\
&= n^r(n) \\
&= n^{r+1} \\
\end{align*}
$$

For the lower bound, let us restrict to the last half of the terms. For all $k$ with $\frac{n}{2} \leq k \leq n$, we have $(\frac{n}{2})^r \leq k^r$, so:
$$
\begin{align*}
\sum_{k=1}^{n}k^r \geq \sum_{k=n/2}^{n}k^r
&\geq \sum_{k=n/2}^{n}\left(\frac{n}{2}\right)^r \\
&= \left(\frac{n}{2}\right)^r(n-\frac{n}{2}+1) \\
&\geq \left(\frac{n}{2}\right)^r\left(\frac{n}{2}\right) \\
&= \left(\frac{n}{2}\right)^{r+1} \\
&= \frac{1}{2^{r+1}}n^{r+1} \\
\end{align*}
$$

Putting it altogether, we can bound the original sum:
$$
\begin{align*}
\frac{1}{2^{r+1}}n^{r+1} \leq \sum_{k=1}^{n}k^r \leq n^{r+1} \\
\end{align*}
$$

i.e.,
$$
\sum_{k=1}^{n}k^r=\Theta(n^{r+1}) \\
$$

### Method 2

Since $f(x)=x^r$ is monotonically increasing for $r\geq 0$, we can use approximation by integrals:
$$
\begin{align*}
\int_{1}^{n+1} x^r\,dx \leq \sum_{k=1}^{n}k^r \leq \int_{0}^{n} x^r\,dx \\
\end{align*}
$$

Now evaluate the integrals:
$$
\begin{align*}
\int_{1}^{n+1} x^r\,dx
&= \left[\frac{x^{r+1}}{r+1}\right]_{1}^{n+1} \\
&= \frac{(n+1)^{r+1}-1}{r+1} \\
\int_{0}^{n} x^r\,dx
&= \left[\frac{x^{r+1}}{r+1}\right]_{0}^{n} \\
&= \frac{n^{r+1}}{r+1} \\
\end{align*}
$$

Therefore:
$$
\begin{align*}
\frac{(n+1)^{r+1}-1}{r+1} \leq \sum_{k=1}^{n}k^r \leq \frac{n^{r+1}}{r+1} \\
\end{align*}
$$

i.e.,
$$
\sum_{k=1}^{n}k^r=\Theta(n^{r+1}) \\
$$

<h3 style="display:inline">(10)</h3> give asymptotically tight bounds for $\displaystyle\sum_{k=1}^{n}lg^kk$

Let $a_k=(\lg k)^k$. For $k\geq 3$, trivially, we have $\lg(k+1)\geq 2$ and $\lg k\leq \lg(k+1)$, so:
$$
\begin{align*}
\frac{a_k}{a_{k+1}}
&= \frac{(\lg k)^k}{(\lg(k+1))^{k+1}} \\
&= \frac{(\lg k)^k}{(\lg(k+1))^k}(\frac{1}{\lg(k+1)}) \\
&= \left(\frac{\lg k}{\lg(k+1)}\right)^k(\frac{1}{\lg(k+1)}) \\
&\leq 1^k\cdot \frac{1}{\lg(k+1)} \\
&\leq 1\cdot \frac{1}{2} \\
&= \frac{1}{2} \\
\end{align*}
$$

Thus, for all $k\geq 3$:
$$
a_k \leq \frac{1}{2}a_{k+1}
$$

Iterating this inequality, for any $k$ with $3\leq k\leq n$ we have:
$$
\begin{align*}
a_k 
&\leq \frac{1}{2}a_{k+1} \\
&\leq \frac{1}{2^2}a_{k+2} \\
&\;\;\vdots \\
&\leq \left(\frac{1}{2}\right)^{n-k}a_n \\
\end{align*}
$$

For the upper bound:
$$
\begin{align*}
\sum_{k=1}^{n}(\lg k)^k
&= a_1+a_2+\sum_{k=3}^{n}a_k \\
&\leq a_1+a_2+\sum_{k=3}^{n}\left(\frac{1}{2}\right)^{n-k}a_n \\
&= a_1+a_2+a_n\sum_{k=3}^{n}\left(\frac{1}{2}\right)^{n-k} \\
&= a_1+a_2+a_n\sum_{j=0}^{n-3}\left(\frac{1}{2}\right)^j && \text{(Reindex } j=n-k\text{)} \\
&\leq a_1+a_2+a_n\sum_{j=0}^{\infty}\left(\frac{1}{2}\right)^j \\
&= a_1+a_2+a_n\left(\frac{1}{1-\frac{1}{2}}\right) && \text{(Appendix A.7)} \\
&= a_1+a_2+2a_n \\
&= a_1+a_2+2(\lg n)^n \\
\end{align*}
$$

For the lower bound, the sum is at least its largest term (the $k=n$ term):
$$
\begin{align*}
\sum_{k=1}^{n}(\lg k)^k \geq (\lg n)^n \\
\end{align*}
$$

Putting it altogether:
$$
\begin{align*}
(\lg n)^n \leq \sum_{k=1}^{n}(\lg k)^k \leq a_1+a_2+2(\lg n)^n \\
\end{align*}
$$

i.e., 
$$
\begin{align*}
\sum_{k=1}^{n}\lg^k k=\Theta((\lg n)^n) \\
\end{align*}
$$
