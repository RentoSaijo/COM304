Recall that a <b>problem</b> is a generalized form of a specific problem that must be solved for all possible inputs.  

This means the inputs can be variable in both size and also in terms of numeric domain if not specified up to some fixed but asympototically large value <b>n</b>. 

<b><u>Input:</u></b> A __[sequence](https://en.wikipedia.org/wiki/Sequence)__ of $n$ numbers $A = \langle a_1,a_2, ... ,a_n \rangle$

<b><u>Output:</u></b> A __[permutation](https://en.wikipedia.org/wiki/Permutation)__ $A' = \langle a'_1,a'_2, ... ,a'_n \rangle : a'_1 \leq a'_2\leq ... \leq a'_n$

The individual numbers in $A$ are called <b>keys</b> and $\forall a \in A \wedge a \in \mathbb{Z^+} $

We will now use __[pseudocode](https://en.wikipedia.org/wiki/Pseudocode)__ from the text to detail our first __[algorithm](https://en.wikipedia.org/wiki/Algorithm)__ which is a correct general solution to a generalized problem.  In this case we will look at the well know algorithm __[Insertion Sort](https://en.wikipedia.org/wiki/Insertion_sort)__ 


Compare the above to actual python3 code for insertion sort below:

```python
#instance code----------------------------------------------------------------------------------------------
import numpy as np                                                        #import numpy package
n = int(np.random.sample()*100)+100                                       #random positive n value 100 to 200
A = np.random.choice([int(np.random.sample()*100)+1 for i in range(n)],n) #now we set each value to 1 to 100

#algorithm code---------------------------------------------------------------------------------------------
def insert_sort(A):
    for j in range(0, len(A) - 1)[::-1]:
        k = A[j]
        i = j+1
        while i<len(A) and A[i]>k:
            A[i-1] = A[i]
            i = i+1
        A[i-1] = k
    return A
print(A)
print(insert_sort(A))
```

### Loop Invariant and Correctness

A __[loop invariant](https://en.wikipedia.org/wiki/Loop_invariant)__ is a analysis device that helps to prove correctness: the abilty to show that an algorithm is correct for size $n$
It is similair in some ways to __[Math Induction](https://en.wikipedia.org/wiki/Mathematical_induction)__ 

It works by thinking of a portion or part of the input that is underging algorithm manipulation.

At the start of each iteration in the above algorithm, we can consider the first subsequence (sub array, array slice...):  $A[1...j-1]$

We can see thatas the algrthm proceeds, it will not have to touch the first element, since it is the smallest in the list, ...

(a) <u>Initialization</u>: true prior to the step

(b) <u>Maintenance</u>: true before the step, true after the step (most iportant)

(b) <u>Termination</u>: terminates correcty?  (a) and (b) usually $\implies$ (c) 

#### Now apply to INSERTION-SORT:

(a) $j=2$ means $A[1..j-1]$ is true! 

(b) because $i=j-1$ in <b>line 3</b>, we are moving any keys to the left that are greater than the value contained in <b>key</b> in <b>line 4</b>, so this is true!
