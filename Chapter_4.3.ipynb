## 4.3 <u>Substitution Method For Recurrence Relations<u>

This section is all about methods for solving recurrence relations that appear in Divide and Conquer style algorthms.  First we will travel back to 2.3 and look at a very good sorting algorithm called merge sort (unlike insert sort, this one runs very fast!)

### 2.3 <u>Merge Sort Algorithm<u>

![title](https://upload.wikimedia.org/wikipedia/commons/thumb/c/cc/Merge-sort-example-300px.gif/220px-Merge-sort-example-300px.gif)

The [merge sort](https://en.wikipedia.org/wiki/Merge_sort) algorthm works by spliting the input in half and then recursing until a subarray of size 2 remains.  Then at this points they can be compared and swaped as needed using one instruction.  When this baseline is completed the recursive function returns which causes the last two sub arrays that were split to be merged, which in this initial case means that two subarrays of size two each need to be sorted.  The arrays do not need a regular sort here since each array itself is sorted.  This leads to a much faster way of dealing with the merging of two unsorted subarrays.  More specifically we will use the following pseudo code:

in python3 this can be written by copying much of the above and correct for zero indecies and using the np for inf:

```python
#instance code-----------------------------------------------------------------------------------------------
import numpy as np                                                        #import numpy package
n = int(np.random.sample()*100)+100                                       #random positive n value 100 to 200
A = np.random.choice([int(np.random.sample()*100)+1 for i in range(n)],n) #now we set each value to 1 to 100

#algorithm code----------------------------------------------------------------------------------------------
def merge(A,p,q,r):
    n1,n2 = q-p+1,r-q
    L = [A[p+i] for i in range(n1)]  +[np.iinfo(int).max]
    R = [A[q+j+1] for j in range(n2)]+[np.iinfo(int).max]
    i,j = 0,0
    for k in range(p,r+1,1):
        if L[i]<=R[j]: 
            A[k] = L[i]
            i += 1
        else:
            A[k] = R[j]
            j += 1
            
def merge_sort(A,p,r):
    if p<r:
        q = int(np.floor((p+r)/2))
        merge_sort(A,p,q)
        merge_sort(A,q+1,r)
        merge(A,p,q,r)

print(A)
merge_sort(A,0,n-1)
print(A)
```

Section <b>2.3</b> concludes with a resulting calculation of the run time complexity as a [recurrence relation](https://en.wikipedia.org/wiki/Recurrence_relation) shown here:

$T(n) =
  \begin{cases}
    \Theta(1)       & \quad \text{if } n \leq c\\
    aT(n/b)+D(n)+C(n)  & \quad \text{otherwise }
  \end{cases}$

How do we then get the asympototic closed form that we need to quatify how good the algorithm is? (We will look at the two rigorous methods, one is in the section and one is in the next section)
<br><br>

#### <u>Substitution Method</u> (AKA guess and check)

<b><font color="red">(1)</font></b> guess the solution (yes, you are reading this correctly!)

<b><font color="green">(2)</font></b> use math induction (it will either suceed or fail which means you have to start at (1) again)


<b>\[EX 1\]</b> $T(n) = 2T(\lfloor n/2 \rfloor) + n$

<b><font color="red">(1)</font></b> $T(n) = O(n\;lg\;n)$

<b><font color="green">(2)</font></b> $T(n) \leq cn\;lg\;n\;:\;\;$ <b><font color="green">$c>0$</font></b>

<u>proof:</u>

assume it holds $\forall m \in \mathbb{Z}^{+}:\;\;m<n$

let $m = \lfloor  n/2 \rfloor$ then,

$T(\lfloor  n/2 \rfloor) \leq 2(c\lfloor  n/2 \rfloor lg(\lfloor  n/2 \rfloor))+n$

$\;\;\;\;\leq cn\;lg(n/2)+n$

$\;\;\;\;= cn\;lg(n)-cn\;lg(2)+n$

$\;\;\;\;= cnlg(n)-cn+n$

$\;\;\;\;\leq cnlg(n): c \geq 1$ in this case (we have to correct the original <b><font color="green">$c>0$</font></b> in step <b><font color="green">(2)</font></b>) 

$\therefore T(n) = O(n\;lg\;n) \blacksquare$

<br><br>

<b>\[EX 2\]</b> $T(n) = T(n-1) + \Theta(n)$

<b><font color="red">(1 Guess)</font></b> $T(n) = \Theta(n^2)$

<b><font color="green">(2a show $O$)</font></b> $T(n) \leq T(n-1)+an\;:\;\;$ <b><font color="green">$a>0$</font></b>

<b><font color="green">(2b show $\Omega$)</font></b> $T(n) \geq T(n-1)+bn\;:\;\;$ <b><font color="green">$b>0$</font></b>

<u>(2a) proof:</u>

$\;\;$ Assume it holds for <b><font color="red">$T(m) = cm^2$</font></b>$, \; c \in \mathbb{Z}^{+},\;m<n\;\;\;\;$ <b><font color="red">(inductive hypothesis)</font></b>

$\;\;$ Now we must show it will hold for <b><font color="blue">$m = n-1: \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$</font></b> <b><font color="blue">(inductive step)</font></b>

$\;\;$ <b><font color="green">$T(n) \leq T(n-1)+an$</font></b> $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$<b><font color="green">(original recurrence)</font></b>

$\;\;\;\; = $ <b><font color="red">$c(n-1)^2+an$</font></b> $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ <b><font color="red">(now we sub the inductive hypothesis)</font></b>

$\;\;\;\;= c(n-1)(n-1)+an$ $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ (algebra expansion and simplification)

$\;\;\;\;= cn^2-2cn+c+an$ $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ (get to the inductive hypothesis using $a$)

$\;\;\;\;\leq n^2+1: a=2,c=1$

$\;\;\;\; \therefore T(n) = O(n^2) \blacksquare$

<u>(2b) proof:</u>

$\;\;$ Assume it holds for <b><font color="red">$T(m) = dm^2$</font></b>$, \; d \in \mathbb{Z}^{+},\;m<n\;\;\;\;$ <b><font color="red">(inductive hypothesis)</font></b>

$\;\;$ Now we must show it will hold for <b><font color="blue">$m = n-1: \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$</font></b> <b><font color="blue">(inductive step)</font></b>

$\;\;$ <b><font color="green">$T(n) \geq T(n-1)+bn$</font></b> $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$<b><font color="green">(original recurrence)</font></b>

$\;\;\;\; = $ <b><font color="red">$c(n-1)^2+an$</font></b> $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ <b><font color="red">(now we sub the inductive hypothesis)</font></b>

$\;\;\;\;= d(n-1)(n-1)+bn$ $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ (algebra expansion and simplification)

$\;\;\;\;= dn^2-2dn+d+bn$ $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ (get to the inductive hypothesis using $a$)

$\;\;\;\;\leq n^2+1: b=2,d=1$ $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ ($O(1)$ can be removed, but $O(n)$ can't!)

$\;\;\;\; \therefore T(n) = \Omega(n^2) \blacksquare$

By (2a) and (2b) $\implies T(n) = T(n-1)+\Theta(n)\;$ has solution $T(n) = \Theta(n^2)$

#### <u>Pitlfalls</u> (issues you may encounter)

<b>\[EX 3\]</b> $T(n) = 2T(\lfloor n/2 \rfloor)+n$

<b><font color="red">(1)</font></b> $T(n) \leq cn$

<b><font color="green">(2)</font></b> $T(n) \leq 2(c\lfloor n/2 \rfloor) + n$

$\;\;\;\;\leq cn + n$

$\;\;\;\;=O(n) \blacksquare$

We also need to watch out for algebra issues that can arise (due to a bad guess)

Guess is: <font color="red">$T(n) \leq cn^3$</font>

$T(n) \leq c(n-1)^3 + an^2$

$\;\; = c(n-1)(n-1)(n-1)+an^2$

$\;\; = c(n^2-2n+1)(n-1)+an^2$

$\;\; = c(n^3-3n^2+3n-1)+an^2$

$\;\; = cn^3-3cn^2+3cn-c+an^2$

$\;\; = n^3+3n-1: c=1,a=3$ $\;\;\;\;$ ($T(n) \leq cn^3-dn)$ would knock out $3n$)

$\therefore T(n) = O(n^3)$

<b><font color="red"> incorrect!!!</font></b> Induction is off my a term in the guess

#### <u>Change of Variable Trick</u>

<b>\[EX 4\]</b> $T(n) = 2T(\lfloor n/2 \rfloor)+lg\;n$

<u>change:</u> let $m=lg\;n \implies T(2^m) = 2T(2^{m/2})+m$

<u>rename:</u> $S(m) = T(2^m) \implies S(m) = 2S(m/2)+m \implies S(m) = O(m\;lg\;m)$

<u>change:</u> $S(m)\; \text{to} \;T(n) \implies T(n) = T(2^m) = 2T(2^{m/2})+m = S(m) = O(lg\;n\;lg\;lg\;n)$
